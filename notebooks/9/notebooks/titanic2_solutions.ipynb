Predspracovaný dataset obsahuje nasledovné atribúty:

* pclass - trieda, v ktorej pasažier cestoval
* survived - udáva, či pasažier nehodu parníku prežil (1) alebo nie (0)
* sex - pohlavie cestujúceho
* age - vek
* sibsp - počet súrodencov pasažiera, resp. druhov/družiek
* parch - počet rodičov/detí medzi cestujúcimi
* ticket - číslo lístka
* fare - výška cestovného
* cabin - kajuta v ktorej bol cestujúci ubytovaný
* embarked - prístav, kde cestujúci nastúpil (Southampton, Cherbourg, Queenstown)
* family - počet rodinných členov, s ktorými cestujúci cestoval
* has_family - udáva, či cestujúci cestoval s rodinným príslušníkom alebo nie
* fare_ordinal - diskretizácovaná cena
* title - titul cestujúceho
* title_short - skratka titulu
* age_ordinal - stupeň veku - vznikol diskretizáciou atribútu age
* deck - paluba na ktorej bol ubytovaný

#### Úloha 9.18

Vizualizujte korelácie numerických atribútov pomocou Heatmapy. Korelačné koeficienty vykreslite do grafu. 

```python
# YOUR CODE HERE

g = sns.heatmap(titanic.corr(),annot=True, cbar=False, cmap="Blues")

# vykreslime heatmpapu na korelacnej tabulke datoveho ramca titanic (titanic.corr())
# anotacie (annot) nastavime na True (vypisu sa koeficienty)
# vypli sme legendu (cbar = False)
# nastavili sme vhodnu farebnu paletu, kde farby vhodne doplnaju koeficienty (cmap="Blues")
```

#### Úloha 9.19

Vyberte si niektorú z konfigenčných tabuliek z úloh z predošlého cvičenia a vykreslite ju pomocou heatmapy.

```python
# YOUR CODE HERE

# pouzijeme kontigencnu tabulku z ulohy 8.5
table = pd.pivot_table(titanic, index="embarked", columns="fare_ordinal", values="survived", aggfunc="mean", margins=True)
g = sns.heatmap(table, annot=True, cbar=False, cmap="Blues")

# tabulku si ulozime do premennej table
# tuto potom mozeme pouzit ako zdroj dat pre heatmapu
# vykreslime heatmapu rovnako ako v predchadzajucej ulohe
```
